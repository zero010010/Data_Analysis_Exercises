```python
import requests 
import pandas as pd

URL = "https://swapi.dev/api/" #URL de la API   
llamada=requests.get(URL)#Obtener la respuesta de la URL
URL_LIST=llamada.json()#Convertir la respuesta a JSON

for clave in URL_LIST.keys():# iterar sobre las claves del diccionario(lista)
    llamada=requests.get(URL_LIST[clave])#Obtener la respuesta de la URL, llamada API
    df=pd.DataFrame(llamada.json()['results']) #Convertir la respuesta a dataframe JSON
    df.to_csv(f"{clave}.csv",index=False) # Guardar el dataframe en un archivo CSV


    
```
