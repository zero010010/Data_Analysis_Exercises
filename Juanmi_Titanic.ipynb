### 34. Usando [`kdeplot`]("https://seaborn.pydata.org/generated/seaborn.kdeplot.html") de `seaborn` represente los que no sobrevivieron distribuidos por edad.

Sugerencia: Primero debes "filtrar" el conjunto de datos `titanic` donde la columna "survived" es 0, indexando la columna `"age"` solamente.

Argumentos que debes pasar a la función: 

    - color = "rojo"
    - label = "No sobrevivido"
    - shade = True
    
Deberías obtener algo como esto:

![kdeplot](./img/kdeplotsur.png)

```python
sns.kdeplot(data=titanic[titanic['survived']==0], x="age", color="red", label="Not Survived", shade=True);
plt.legend();
```

```python
sns.kdeplot(data=titanic[titanic['survived']==0], x="age", color="red", label="Not Survived")
sns.histplot(data=titanic[titanic['survived']==0], x="age", color="red", alpha=0.3, bins=20)
sns.kdeplot(data=titanic[titanic['survived']==1], x="age", color="green", label="Survived")
sns.histplot(data=titanic[titanic['survived']==1], x="age", color="green", alpha=0.3, bins=20)
plt.legend()
```

```python
sns.kdeplot(data=titanic, x="age", hue='survived', shade=True);
```
